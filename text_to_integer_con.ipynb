```python
import os

def replace_text_with_int(input_root, output_root, mapping, file_ext=".txt"):
    """
    Replace specific text labels with integers in ALL files across ALL folders.
    """
    # Walk through all subdirectories
    for subdir, _, files in os.walk(input_root):
        for file in files:
            if file.lower().endswith(file_ext):  # handle .txt or .TXT
                input_path = os.path.join(subdir, file)

                # Mirror folder structure inside output_root
                relative_path = os.path.relpath(subdir, input_root)
                out_dir = os.path.join(output_root, relative_path)
                os.makedirs(out_dir, exist_ok=True)
                output_path = os.path.join(out_dir, file)

                # Read original file
                with open(input_path, "r") as f:
                    lines = f.readlines()

                # Replace text → int
                new_lines = []
                for line in lines:
                    new_line = line
                    for text, num in mapping.items():
                        if text in new_line:
                            new_line = new_line.replace(text, str(num))
                    new_lines.append(new_line)

                # Save to new location
                with open(output_path, "w") as f:
                    f.writelines(new_lines)

                print(f"✅ Processed: {input_path} -> {output_path}")


# Example usage
input_root = ("/content/drive/MyDrive/lab5")
output_root = ("/content/drive/MyDrive/0000")
mapping = {
    "using phone": 0,
    "smoking": 1,
    "eating": 2
}

replace_text_with_int(input_root, output_root, mapping)
```
